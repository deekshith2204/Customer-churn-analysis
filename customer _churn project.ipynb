```python
#simple linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
```

```python
x=pd.DataFrame(customer_df.loc[:,'tenure'])
y=pd.DataFrame(customer_df.loc[:,'MonthlyCharges'])
```

```python
x.shape
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
linearregression=LinearRegression()
```

```python
model=linearregression.fit(x_train,y_train)
```

```python
predict=linearregression.predict(x_test)
```

```python
predict=pd.DataFrame(predict,columns=['predict'])
```

```python
r2_score(y_test,predict)
```

```python
mse=mean_squared_error(y_test,predict)
```

```python
rmse=np.sqrt(mse)
rmse
print('the root of mean squared error is :',rmse)
```

```python
x=pd.DataFrame(customer_df['MonthlyCharges'])
num=pd.get_dummies(customer_df['Churn'])
y=pd.DataFrame(num['Yes'])
```

```python
#simple logistic model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score

logistic= LogisticRegression(max_iter=500)
model=logistic.fit(x_train,y_train)
```

```python
predict=logistic.predict(x_test)
predict=pd.DataFrame(predict,columns=['predicted'])
```

```python
accuracy_score(y_test,predict)
```

```python
confusion_matrix(y_test,predict)
```
